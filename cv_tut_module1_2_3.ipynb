```python
import matplotlib.pyplot as plt # used for plot out and display images inside the notebook
%matplotlib inline # isn't required anymore
```

```python
from PIL import Image
```

```python
pic = Image.open('drive/MyDrive/Colab Notebooks/DATA/00-puppy.jpg')
```

```python
type(pic)
```

```python
pic_arr = np.asarray(pic)
```

```python
pic_arr
```

```python
type(pic_arr)
```

```python
pic_arr.shape # 1300 x 1950 pixels, by 3 color channels
```

```python
plt.imshow(pic_arr) # used to show numpy arrays of pictures as pictures
```

```python
pic_red = pic_arr.copy()
```

```python
plt.imshow(pic_red)
```

```python
# R G B, channels are in order R = 0, G = 1, B = 2

# RED Channel values, 0-255, #cmap = color mapping, gray = grayscale.
# the closer to white the more red pixels there are.
plt.imshow(pic_red[:,:,0],cmap='gray')
```

```python
plt.imshow(pic_red[:,:,1],cmap='gray')
```

```python
plt.imshow(pic_red[:,:,2],cmap='gray')
```

```python
pic_red[:,:,1] = 0
pic_red[:,:,2] = 0
```

```python
plt.imshow(pic_red)
```
